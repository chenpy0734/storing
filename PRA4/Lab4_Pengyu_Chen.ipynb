## 1. Open the data, perform basic data preparation

```python
# import modules
import numpy as np
import pandas as pd
```

#### 1.1 Read and prepare composition vs temperature data.

```python
# read and prepare composition vs temperature data
compo_temp = pd.read_csv('VO2 - Nb2O3 Composition and temp Combiview.txt', delimiter='\t')
compo_temp = compo_temp.dropna()#drop nan
compo_temp.drop_duplicates()#drop duplicates
print(compo_temp.shape)
compo_temp
```

#### 1.2 Read and prepare XRD data.

```python
# read and prepare XRD data
xrd = pd.read_csv('VO2 -Nb2O3 XRD Combiview.txt', delimiter='\t')
xrd = xrd.dropna()#drop nan
xrd.drop_duplicates()#drop duplicates
print(xrd.shape)
xrd
```

## 2. Plot and navigate the datasets.

#### 2.a Generate composition vs. temperature plot and show points clearly.

```python
#import modules to plot
import matplotlib.pyplot as plt

#choose the figure size
plt.figure(figsize=(9,9))

#plot composition vs temperature with label
plt.title('Composition vs Temperature', fontsize=18)
plt.ylabel('Temperature (C)', fontsize=11)
plt.xlabel('Composition of V (at%)', fontsize=11)

#set the y limit with 5 degree interval
plt.ylim(20, 70)
plt.yticks(np.arange(20, 70, 5))

#set the x limit with 5 degree interval
plt.xlim(50, 100)
plt.xticks(np.arange(50, 100, 5))

plt.grid(True)#add grid

# plot points
plt.plot(compo_temp['V'], compo_temp['temp'], 'ko', markersize=5)
```

 #### 2.b Generate an x-ray diffraction vs temperature plot for a constant V composition of 99at%.

```python
## plot xrd parttern for V composition of 99 at%
index_list_v99 = []#create a list to store the index of the composition of 99 at%
temperature_list_v99 = []#create a list to store the temperature of the composition of 99 at%

#for loop to append the index of the composition of 99 at% into list
for i in range(len(compo_temp["V"])):
    if compo_temp["V"][i] == 99:
        index_list_v99.append(i)
        temperature_list_v99.append(compo_temp["temp"][i])

#use the index to extreact corresponding xrd data (this is y value)
xrd_v99 = xrd.iloc[index_list_v99]

#extract colume name, change from string to float and store in a list (this is x value)
b2_x_list = list(xrd_v99.columns)
for i in range(len(b2_x_list)):
    b2_x_list[i] = float(b2_x_list[i])


#plot xrd pattern for V composition of 99 at%
#Preapre the pattern size and information
plt.figure(figsize=(20,10))
plt.title('XRD Pattern for V Composition of 99 at% @ Different Temperature', fontsize=18)
plt.xlabel('2Theta (degree)', fontsize=11)
plt.ylabel('Intensity', fontsize=11)

#set x and y limit
plt.xlim(left=17, right=38)
plt.xticks(np.arange(17, 38))
plt.ylim(bottom=20, top=150)
plt.yticks(np.arange(20, 150, 25))
plt.grid(False)

#plot xrd pattern for V composition of 99 at% with label
for i in range(len(xrd_v99)):
    plt.plot(b2_x_list, xrd_v99.iloc[i], label = str(temperature_list_v99[i]) + ' degreeC')###copilot help to generate label###
plt.legend(loc='upper right', fontsize=11)

#print heat map for V composition of 99 at%
xrd_v99_hm = xrd_v99.to_numpy()
plt.figure()
plt.pcolor([x for i, x in enumerate(xrd_v99_hm) if i not in []])
plt.ylabel('Material Points')
```

#### 2.c Generate an X-ray diffraction vs composition plot for a constant temperature of 23 degreeC.

```python
## plot xrd parttern for temperature of 23 degreeC
index_list_dc23 = []#create a list to store the index of the temperature of 23 degreeC
composition_list_dc23 = []#create a list to store the composition of the temperature of 23 degreeC

#find the index of the temperature of 23 degreeC and store in list
for i in range(len(compo_temp["temp"])):
    if compo_temp["temp"][i] == 23:
        index_list_dc23.append(i)
        composition_list_dc23.append(compo_temp["V"][i])

#use the index to extreact corresponding xrd data (this is y value)
xrd_dc23 = xrd.iloc[index_list_dc23]

#extract colume name, change from string to float and store in a list (this is x value)
c2_x_list = list(xrd_dc23.columns)
for i in range(len(c2_x_list)):
    c2_x_list[i] = float(c2_x_list[i])

#plot xrd pattern for temperature of 23 degreeC
#Preapre the pattern size and information
plt.figure(figsize=(20,10))
plt.title('XRD Pattern for Temperature of 23 degreeC @ Different Composition', fontsize=18)
plt.xlabel('2Theta (degree)', fontsize=11)
plt.ylabel('Intensity', fontsize=11)

#set x and y limit
plt.xlim(left=17, right=38)
plt.xticks(np.arange(17, 38))
plt.ylim(bottom=15, top=150)
plt.yticks(np.arange(15, 150, 25))
plt.grid(False)

#plot xrd pattern for temperature of 23 degreeC with label
for i in range(len(xrd_dc23)):
    plt.plot(c2_x_list, xrd_dc23.iloc[i], label = str(composition_list_dc23[i]) + ' at%')###copilot help to generate label###
#locate the legend at the bottom right
plt.legend(ncol=2, loc='upper right', fontsize=11)###copilot help to generate two columns of label###

#print heat map for temperature of 23 degreeC
xrd_dc23_hm = xrd_dc23.to_numpy()
plt.figure()
plt.pcolor([x for i, x in enumerate(xrd_dc23_hm) if i not in []])
plt.ylabel('Material Points')
```

## 3. Perform some form of dimensional reduction on the diffraction data.

#### 3.a PCA of XRD points.

```python
#impory modules for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#prepare data for PCA
xrd_to_array = xrd.to_numpy()#convert xrd data to array
xrd_to_array
n_comp = min(len(xrd_to_array[:,0]), len(xrd_to_array[0,:]))#check the n components

#conduct PCA, refer to PRA syntax
pca = PCA(n_components=n_comp)
pca.fit(xrd_to_array)
loading_vectors = pca.components_#loading eigenvectors of covariance matrix
explained_ variance = pca.explained_variance_#loading variance
xrd_reduced = pca.transform(xrd_to_array)#reduced data

print(xrd_reduced.shape)#check the shape of reduced data
xrd_reduced#pca analysis result

#PCA plot and corresponding information
#plt.figure(figsize=(8,6))
#plt.plot(xrd_reduced[:,0],xrd_reduced[:,1],'k.')#plot pca
#plt.xlabel('PCA 1', fontsize=11)
#plt.ylabel('PCA 2', fontsize=11)
```

```python
#use screen plot to get the variance for each component#

#refer to PRA syntax screen plot
PC_values = np.arange(n_comp) + 1#set the x axis

#plot the variance for each component
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')

#Table information
plt.title('Scree Plot', fontsize=18)
plt.xlabel('Principal Component', fontsize=11)
plt.ylabel('Variance Explained', fontsize=11)
plt.show()
```

```python
#plot the Explained Variance Ratio
#refer to PRA syntax
PC_values = np.arange(n_comp) + 1
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('Explained Variance Plot',fontsize=18)
plt.xlabel('Principal Component', fontsize=11)
plt.ylabel('Total Variance Explained', fontsize=11)
plt.xlim(0,400)
plt.show()
```

#### 3.b USe total explained variance to drop noise


```python
#set a threshold for the explained variance ratio

#insult ChatGPT from here
threshold = 0.95
n_comp_threshold = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= threshold) + 1
print(n_comp_threshold)
#to here#################

#conduct PCA again with the threshold
pca2 = PCA(n_components=n_comp_threshold)
pca2.fit(xrd_to_array)

#transform the data
xrd_reduced_threshold = pca2.transform(xrd_to_array)
xrd_reduced_threshold
```

## 5. Perform clustering analyses to get similar phase diagram.

```python
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import pairwise_distances
import numpy.matlib as nm

inertias = {}

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(xrd_reduced_threshold)
    inertias[k] = kmeans.inertia_ 
    

plt.figure();
plt.plot(list(inertias.keys()), list(inertias.values()));
plt.title('K-Means, Elbow Method')
plt.xlabel("Number of clusters, K");
plt.ylabel("Inertia");
```

```python
k = 4
kmeans = KMeans(n_clusters=k, random_state = 0).fit(xrd_reduced_threshold)
L = list(kmeans.labels_)
# assign colours to the labels
clist = ['r','g','b','k','c','m','y'] # for up to 7 colors
for cl in range(k):
    L = [clist[cl] if i==cl else i for i in L]
    
cluster_centers = kmeans.cluster_centers_
plt.figure(figsize = (8,6))
plt.subplot(1,1,1)
# plt.cla()
plt.scatter(xrd_reduced_threshold[:,0],xrd_reduced_threshold[:,1],c = L)
plt.plot(cluster_centers[:,0], cluster_centers[:,1], 'rx', markersize=7)
plt.title('k-means results')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
# find the distance between the cluster centers and the cluster points
d = pairwise_distances(nm.concatenate((cluster_centers, xrd_reduced_threshold), axis=0))

# plot the cluster centers and indicate which point is closest to each center.
plt.figure(1, figsize = (8,16))
plt.subplot(k+1,1,1)
plt.scatter(xrd_reduced_threshold[:,0],xrd_reduced_threshold[:,1],c =L)
plt.plot(cluster_centers[:,0],cluster_centers[:,1],'kx',markersize=10)

# find the cluster points closest to the cluster centers
i = 1
while i < (k+1):
    cluster_rep = nm.argmin(d[i-1,k:],axis = 0)
    plt.plot(xrd_reduced_threshold[cluster_rep,0], xrd_reduced_threshold[cluster_rep,1],'r.')
    plt.annotate(str(L[cluster_rep]), (xrd_reduced_threshold[cluster_rep,0], xrd_reduced_threshold[cluster_rep,1]))
    i+=1
plt.xlabel('PC1')
plt.ylabel('PC2')    
```

```python
# plot the XRD patterns of the cluster points closest to the cluster centers
plt.figure(1, figsize = (8,20))
i = 1
while i < (k+1):
    cluster_rep = nm.argmin(d[i-1,k:],axis = 0)
    plt.subplot(k+1,1,i+1)
    plt.plot(xrd_to_array[cluster_rep,:])
    plt.title('Cluster ' + str(i))
    plt.xlabel('2$\Theta$')
    plt.ylabel('Intensity')
    i+=1
```
